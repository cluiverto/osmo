```python
from datasets import load_dataset

ds = load_dataset("seyonec/goodscents_leffingwell")
ds
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Assuming ds is already defined and contains the graph data
graph = ds["train"][123]

edges = graph["edge_index"]
num_edges = len(edges[0])
num_nodes = graph["num_nodes"]

# Conversion to NetworkX format
G = nx.Graph()
G.add_nodes_from(range(num_nodes))
G.add_edges_from([(edges[0][i], edges[1][i]) for i in range(num_edges)])

# Set up the plot
plt.figure(figsize=(10, 8))  # Set figure size

# Define a layout for our nodes 
pos = nx.spring_layout(G)  # You can also try nx.circular_layout(G) or nx.kamada_kaway_layout(G)

# Draw the nodes and edges
nx.draw_networkx_nodes(G, pos, node_size=700, node_color='lightblue', alpha=0.7)
nx.draw_networkx_edges(G, pos, width=2, alpha=0.5, edge_color='gray')

# Draw labels for the nodes
labels = {i: str(i) for i in range(num_nodes)}  # Label nodes with their indices
nx.draw_networkx_labels(G, pos, labels, font_size=12)

# Title and axis settings
plt.title("Graph Visualization", fontsize=16)
plt.axis('off')  # Turn off the axis

# Show the plot
plt.show()
```
