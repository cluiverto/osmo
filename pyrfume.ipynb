```python
# Funkcja do pobierania danych z Pyrfume
def load_pyrfume_data(dataset_name='leffingwell', min_samples_per_class=5):
    """
    Ładuje dane z biblioteki Pyrfume i przygotowuje je do modelu GNN.
    
    Args:
        dataset_name: Nazwa zbioru danych z Pyrfume ('leffingwell', 'goodscents', 'sigma' itp.)
        min_samples_per_class: Minimalna liczba przykładów wymagana dla każdej klasy zapachu
        
    Returns:
        molecules: Lista obiektów OdorMolecule
        odor_classes: Słownik mapujący indeksy klas na ich nazwy
    """
    print(f"Ładowanie danych z Pyrfume ({dataset_name})...")
    
    # Pobranie danych z Pyrfume
    try:
        # Dla zbioru Leffingwell
        if dataset_name == 'leffingwell':
            molecules_df = pyrfume.load_data('leffingwell/molecules')
            odors_df = pyrfume.load_data('leffingwell/odors')
            
            # Łączenie danych o molekułach i zapachach
            merged_df = molecules_df.join(odors_df)
            
            # Usunięcie wierszy z brakującymi wartościami
            merged_df = merged_df.dropna(subset=['smiles', 'odor_description'])
            
            # Ekstrakcja głównej kategorii zapachu (pierwsze słowo opisu)
            merged_df['odor_category'] = merged_df['odor_description'].apply(
                lambda x: x.split(',')[0].strip().lower() if isinstance(x, str) else 'unknown'
            )
            
        # Dla zbioru Goodscents
        elif dataset_name == 'goodscents':
            molecules_df = pyrfume.load_data('goodscents/molecules')
            odors_df = pyrfume.load_data('goodscents/odors')
            
            # Łączenie danych
            merged_df = pd.merge(molecules_df, odors_df, left_index=True, right_index=True)
            merged_df = merged_df.dropna(subset=['smiles', 'odor_description'])
            
            # Ekstrakcja głównej kategorii zapachu
            merged_df['odor_category'] = merged_df['odor_description'].apply(
                lambda x: x.split(',')[0].strip().lower() if isinstance(x, str) else 'unknown'
            )
            
        # Dla zbioru Sigma Aldrich
        elif dataset_name == 'sigma':
            merged_df = pyrfume.load_data('sigma/odors')
            merged_df = merged_df.dropna(subset=['smiles', 'descriptor'])
            merged_df['odor_category'] = merged_df['descriptor']
            
        else:
            raise ValueError(f"Nieznany zbiór danych: {dataset_name}")
            
        # Liczenie wystąpień każdej kategorii i filtrowanie rzadkich zapachów
        category_counts = merged_df['odor_category'].value_counts()
        common_categories = category_counts[category_counts >= min_samples_per_class].index.tolist()
        
        print(f"Znaleziono {len(common_categories)} kategorii zapachów z co najmniej {min_samples_per_class} przykładami każda.")
        
        # Filtrowanie do tylko częstych kategorii
        filtered_df = merged_df[merged_df['odor_category'].isin(common_categories)]
        
        # Mapowanie kategorii na indeksy liczbowe
        category_to_index = {cat: idx for idx, cat in enumerate(common_categories)}
        
        # Tworzenie odwrotnego mapowania dla wyświetlania
        odor_classes = {idx: cat for cat, idx in category_to_index.items()}
        
        # Konwersja danych do formatu OdorMolecule
        molecules = []
        for _, row in filtered_df.iterrows():
            try:
                odor_idx = category_to_index[row['odor_category']]
                mol = OdorMolecule(row['smiles'], odor_idx, row['odor_category'])
                if mol.mol is not None:  # Sprawdzenie czy konwersja SMILES powiodła się
                    molecules.append(mol)
            except Exception as e:
                # Pomiń problematyczne struktury
                continue
                
        print(f"Załadowano {len(molecules)} poprawnych struktur molekularnych.")
        return molecules, odor_classes
        
    except Exception as e:
        print(f"Błąd podczas ładowania danych z Pyrfume: {e}")
        # Zwróć przykładowe dane jeśli wystąpił błąd
        return create_dummy_data()

def create_dummy_data():
    """Tworzy przykładowe dane w przypadku problemów z Pyrfume"""
    print("Tworzenie przykładowych danych...")
    
    data = [
        # Olejek lawendowy
        OdorMolecule("CC(=O)OC1CC(C)(C)C2CCC1(C)C2", 0, "floral"),
        OdorMolecule("CC1=CCC(CC1)C(=C)C", 0, "floral"),
        
        # Cytrusowy
        OdorMolecule("CC(=CCCC(=CC=O)C)C", 1, "citrus"),
        OdorMolecule("CC(=CCCC(C)(C)O)C", 1, "citrus"),
        
        # Drzewny
        OdorMolecule("CC(C)C1=CC(=C(C=C1)O)C(C)C", 2, "woody"),
        OdorMolecule("CC1=CC=C(C=C1)C(C)(C)C", 2, "woody"),
        
        # Miętowy
        OdorMolecule("CC1CCC(C(C1)O)(C)C", 3, "minty"),
        OdorMolecule("CC(C)C1CCC(C)CC1O", 3, "minty"),
        
        # Kwiatowy
        OdorMolecule("COC1=CC=C(C=C1)CC=O", 4, "sweet"),
        OdorMolecule("CC=CC(=O)OC1=CC=CC=C1", 4, "sweet")
    ]
    
    odor_classes = {
        0: "floral",
        1: "citrus",
        2: "woody",
        3: "minty",
        4: "sweet"
    }
    
    return data, odor_classes
```
